```python
new_df = df[['track_name','track_artist','track_album_name','track_album_release_date','playlist_name','playlist_genre']]
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 10000)
```

```python
vectors=cv.fit_transform(new_df['track_name']).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=False,key=lambda x:x[1])
```

```python
new_df.columns = new_df.columns.str.strip()
print(new_df.head())
```

```python
def recommend(music):
    music = music.strip().lower()
    
    new_df.loc[:, 'track_name'] = new_df['track_name'].str.strip().str.lower()
    
    if music in new_df['track_name'].values:
        music_index = new_df[new_df['track_name'] == music].index[0]
        distances = similarity[music_index]
        
        music_list = sorted(list(enumerate(distances)), reverse=False, key=lambda x: x[1])[1:6]
        
        recommended_tracks = [new_df.iloc[i[0]]['track_name'] for i in music_list]
        
        return recommended_tracks
    else:
        return f"'{music}' not found in the dataset."
```

```python
recommend('qué pena')
```
